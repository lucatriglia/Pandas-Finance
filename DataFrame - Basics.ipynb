### Label-based Indexing and Slicing with loc[]
- You would use .loc[] when selecting data based on labels (row names or column names), whereas .iloc[] is for position-based indexing.
- With .loc[] You are selecting rows by index label (not position).

```python
summer
```

```python
summer.loc["DRIVAS, Dimitrios"]
```

```python
summer.loc["PHELPS, Michael"]
```

- As you can see our INDEX contains duplicates. This is allowed in DataFrames
- There is no obligation to have unique values in the index
- BUT BEST PRACTICE IS TO HAVE A CLEAR IDENTIFIER AS IT IS BEST PRACTICE 
- There are limitations to have non-unique indexes

### Label-based Indexing/Slicing Rows and Columns with loc[]

```python
summer.loc["PHELPS, Michael", "Medal"]
```

```python
summer.loc["PHELPS, Michael", ["Medal", "Event"]]
```

```python
summer.loc[["PHELPS, Michael", "LEWIS, Carl"], ["Medal", "Event"]]
```

```python
summer.loc[:, ["Medal", "Event"]] # all rows
```

```python
summer.loc[:"CHASAPIS, Spiridon"] #get all rows until CHASAPIS, Spiridon 
#NOTE THAT AS COMPARED TO LOC, THE LAST VALUE IS INCLUDED
```

```python
summer.loc[:"PHELPS, Michael"]
```

- The above returns an error because PHELPS, Michael is not unique!!
- Until what occurrance should Python slice? 
- This is one of the limitations of having a non-unique index

### Indexing and Slicing with reindex()
- generally used when dealing with timeseries data

```python
import pandas as pd
summer = pd.read_csv("summer.csv")
summer
```

```python
summer.loc[[0,5,3000], ["Athlete", "Medal"]] # DO NOT MIX THIS UP WITH ILOC! THIS IS JUST BECAUSE THE INDEX NAMES ARE ACTUALLY INTS
```

```python
summer.reindex(index = [0,5,30000, 400000], columns = ["Athlete", "Medal"])
```

- **reindex** --> It tries to select rows with the labels 0, 5, 30000, and 400000.
- If these indices don’t exist in the DataFrame, pandas will add NaN (Not a Number) for those rows.

### Adding new Rows .concat()

```python
dic = {"Player": ["messi", "ronaldo", "chiesa"], "Nationality" : ["argentina", "portugal", "italy"], "Club" : ["PSG", "Saudi", "Liverpool"]}
players = pd.DataFrame(data = dic)
players
```

```python
players.loc[3,:] = ["Ramos", "Spain", "Real Madrid"]
players
```

```python
#Adding Many Rows
new = pd.DataFrame(data = [["Salah", "Egypt", "Liverpool"], ["Bonucci", "Italy", "Juventus"]], columns = players.columns)
new
```

```python
players = pd.concat([players, new], ignore_index = True)
players
```

```python
#second example
men2004 = pd.read_csv("men2004.csv")
men2004.head()
```

```python
men2008 = pd.read_csv("men2008.csv")
men2008.head()
```

```python
#without ignore index you repeat some indexes
pd.concat([men2004, men2008], ignore_index = True, keys = None)
```

```python
#however the issue is that we now cannot distinguish between medals in 2008 and 2004
#to get around this we can use the parameter "Keys" to create a secondary index essentially
men0408 = pd.concat([men2004, men2008], ignore_index = False, keys = [2004, 2008], names = ["Year"])
men0408
```

```python
#the above has created a multiIndex...
#these indexes are now TUPLES!
men0408.index
```

```python
men0408.loc[(2004,1),:]
```

```python
#IF WE WANT TO GET OUT OF THIS MULTIINDEX WE JUST SIMPLY RESET INDEX!!
men0408.reset_index()
```

```python
men0408.reset_index().drop(columns = "level_1")
```

### Manipulating Elements in a Dataframe

```python
titanic = pd.read_csv("titanic.csv")
titanic.head()
```

```python
titanic.loc[1,"age"] = 40
titanic.head()
```

```python
titanic.iloc[1,3] = 41
titanic.head()
```

```python
#manipulate many at the same time
titanic.loc[1:3, "age"] = 42
titanic.head()
```

```python
#manipulate many at the same time
titanic.loc[1:3, "age"] = [41,42,43]
titanic.head()
```

```python
#converting age in years to age in months (you must be modifying the same number of rows!)
titanic.loc[:,"age"] = titanic.loc[:,"age"] * 12
titanic.head()
```

```python
#set everyones age to 1 if below 1
titanic.loc[titanic.age < 1, "age"] = 1
```
