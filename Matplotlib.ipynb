# Matplotlib

## Basics

```python
import pandas as pd
titanic = pd.read_csv("titanic.csv")
titanic.head()
```

```python
import matplotlib.pyplot as plt
titanic.info()
```

```python
#only plots NUMERICAL columns
titanic.plot()

#this is the very last method that should be called each time we want to display
plt.show()

#you will see that the x axis is the index #
```

```python
# note that by default it is a lineplot. However, a lineplot is more useful for time series data...
#subplots
titanic.plot(kind = "line", subplots = True, figsize = (15,12), sharex = False)
plt.show()
```

```python
#plotting only a specific column
titanic.age.plot(figsize = (12,8))
plt.show()
```

## Customising Plots

```python
titanic.age.plot(kind = "line", figsize = (12,8), fontsize = 13, c = "red", linestyle = "-", xlim = (0,900), ylim = (0,80))
plt.title("Titanic - Ages", fontsize = 15)
plt.legend(loc = "best", fontsize = 15, shadow = True)
plt.xlabel("Passenger No", fontsize = 13)
plt.ylabel("Age", fontsize = 13)
plt.grid()
plt.show()
```

```python
#the parameters xticks and y ticks take a list of values. So we can create them first
xticks = [x for x in range (0,901,50)]
xticks
```

```python
yticks = [y for y in range (0,81,5)]
```

```python
#note that the grid adjusts automatically
titanic.age.plot(
    kind = "line", figsize = (12,8), fontsize = 13, c = "red", linestyle = "-", 
    xlim = (0,900), ylim = (0,80), 
    xticks = xticks, yticks = yticks,
    rot = 45
)
plt.title("Titanic - Ages", fontsize = 15)
plt.legend(loc = "best", fontsize = 15, shadow = True)
plt.xlabel("Passenger No", fontsize = 13)
plt.ylabel("Age", fontsize = 13)
plt.grid()
plt.show()
```

## Histograms

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
titanic = pd.read_csv("titanic.csv")
titanic.head()
```

```python
titanic.age.value_counts(bins = 10)
```

```python
titanic.age.plot(kind = "hist", figsize = (12,8), bins = 80, fontsize = 15)
plt.show()
```

```python
#this is another way of doing it. there is a direct method called hist()
#the benefit of this is that the documentation is more specific to hist()
#for example in .plot() it is not clear that there is a parameter "bins"
titanic.age.hist(figsize = (12,8), bins = 80, xlabelsize = 15, ylabelsize = 15)
plt.show()
```

```python
#we can also use cumulative parameter
titanic.age.hist(figsize = (12,8), bins = 80, xlabelsize = 15, ylabelsize = 15, cumulative = True)
plt.show()
```

```python
#we can also use density to show relative quantities instead of absolute quantities
titanic.age.hist(figsize = (12,8), bins = 80, xlabelsize = 15, ylabelsize = 15, density = True)
plt.show()
```

## Scatterplots

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
titanic = pd.read_csv("titanic.csv")
titanic.head(4)
```

```python
#you are no longer selecting a single column, you need to specify parameters x and y
titanic.plot(
    kind = "scatter", figsize = (15,8), fontsize = 15, x = "age", y = "fare", xlim = (0,80), ylim = (0,600),
    c = "green", marker = "x", s = 50
)
plt.show()
```

```python
#YOU CAN ALSO PUT ANOTHER COLUMN AS THE COLOR, ESSENTIALLY GIVING YOU A 3D GRAPH
titanic.plot(
    kind = "scatter", figsize = (15,8), fontsize = 15, x = "age", y = "fare", xlim = (0,80), ylim = (0,600),
    c = "pclass", marker = "x", s = 50, colormap = "viridis"
)
plt.show()
```
